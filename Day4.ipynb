```python
#How many different unique email domains names are there in the data set? 591

data['domain'] = data['company email'].str.split('@').str[1]
item_counts = data["domain"].value_counts()
print(item_counts)
```

```python
#Considering only NTAs with at least 5 listed businesses, what is the average total savings and the total jobs created for each NTA?

#Rename NTA column for simplicity
data = data.rename(columns={'Neighborhood Tabulation Area (NTA) (2020)': 'NTA'})

#Filter dataframe to include NTAs that show up at least 5 times then save as dataframe
data_q3 = data.groupby("NTA").filter(lambda x: len(x) > 4)

#Groupby NTA to get mean total savings
mean_savings = data_q3.groupby('NTA')['Total Savings'].mean()

#Groupby NTA to get total jobs created
jobs_created = data_q3.groupby('NTA')['Job created'].sum()
```
